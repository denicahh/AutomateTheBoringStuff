```python
#chapter 10/section 12: video 35: The Raise and Asset Statements (pages 215 - 221)

#raising your own exceptions
#raise Exception('This is the error message.')    #create custome messages 

"""
****************
*              *
*              *
*              *
*              *
****************

"""

def boxPrint(symbol, width, height):
    if len(symbol) !=1:
        raise Exception('"symbol needs to be a string of length 1"')
    if (width < 2) or (height < 2):
        raise Exception('"width and height must be greater or equal to 2."')
    print(symbol * width)
    
    for i in range(height-2):
        print(symbol + (' '* (width-2)) + symbol)
        
    print(symbol * width)
    
boxPrint('*', 15, 5)
boxPrint('O', 5, 16)

#testing the custom exception messages 
#______________________________
#boxPrint('**', 15, 5)
#boxPrint('*', 1, 1)
#______________________________

#The traceback.format_exc() function - traceback format exception function - gets the traceback text from the error as a string value

import traceback
try:
    raise Exception('The is another error message.')
except: 
    errorFile = open('error_log.txt', 'a')
    errorFile.write(traceback.format_exc())
    errorFile.close()
    print('The traceback info was written to error_log.txt')
    
import os
os.getcwd()

#Assertions and the assert statement - assertions are a sanity check to make sure code isn't doing something obviously wrong - assertions are for programmer errors, not user errors. 
#assert False, 'This is the third error message'

market_2nd = {'ns': 'green', 'ew': 'red'}

def switchLights(intersection):
    for key in intersection.keys():
        if intersection[key] == 'green':
            intersection[key] = 'yellow'
        elif intersection[key] == 'yellow':
            intersection[key] = 'red'
        elif intersection[key] == 'red':
            intersection[key] = 'green'
    assert 'red' in intersection.values(), 'Neither light is red!' + str(intersection) #cannot have a green and yellow light at the same time

print(market_2nd)
switchLights(market_2nd)
print(market_2nd)        
    
```

```python
#chapter 10/section 12: video 36: Logging (pages 221 - 225)

#a print function is an example of logging use to debug code

import logging 
logging.basicConfig(level=logging.DEBUG, format='%(asctime)s - %(levelname)s - %(message)s') #setup code for logging in python

#Example: Buggy Factorial Program

#Error version
logging.debug('Start of program')

def factorial(n):
    logging.debug('Start of factorial(%s)' % (n))
    total = 1
    for i in range(n+1):
        total *=i
        logging.debug('i is %s, total is %s' % (i, total))
        
    logging.debug('Return value is %s' % (total))
    return total
#logging.debug('End of program')
print(factorial(5))


#Fixed version
logging.debug('Start of program')

def factorial(n):
    logging.debug('Start of factorial(%s)' % (n))
    total = 1
    for i in range(1, n+1): #range begins at 0 when you just give it one argument 
        total *=i
        logging.debug('i is %s, total is %s' % (i, total))
        
    logging.debug('Return value is %s' % (total))
    return total

print(factorial(5))

#better to use this than print and can turn off all logging messages by disabling rather than having to go back through code and removing all print lines

#Disabled logging version - this version doesn't print any of the logging messages 
logging.disable(logging.CRITICAL)  # this disables all messages from critical level or lower and since critical is the highest it disables them all
logging.debug('Start of program')

def factorial(n):
    logging.debug('Start of factorial(%s)' % (n))
    total = 1
    for i in range(1, n+1): #range begins at 0 when you just give it one argument 
        total *=i
        logging.debug('i is %s, total is %s' % (i, total))
        
    logging.debug('Return value is %s' % (total))
    return total

print(factorial(5))


#5 log levels = debug (lowest) / info / warning / error / critical (highest) can also do logging.info or logging.warning etc 

#can also write to a log file by changing the basic config line and adding a file name to save to - will write to cwd 
logging.basicConfig(filename='myProgramLog.txt', level=logging.DEBUG, format='%(asctime)s - %(levelname)s - %(message)s') #setup code for logging in python
```
