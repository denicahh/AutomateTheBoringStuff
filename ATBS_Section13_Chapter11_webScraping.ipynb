```python
#chapter 11/section 13: video 39: Downloading from the web with the requests module (pages 237 - 240 )

#requests module lets you download files from the web - this is not a standard python module so will need to be installed before it can run
# run this in anaconda prompt : conda install -c anaconda requests


import requests

res = requests.get('https://automatetheboringstuff.com/files/rj.txt')  #returns a response object, stored in a variable it returns the response the site gives
res.status_code  #check if successful if it returns 200 that means it was successful 

len(res.text) #res.text stores the text of the result of the url provided 

#print(res.text[:500])  #return the first 500 characters 


#simpler way to check for success is to call the raise for status method
res.raise_for_status()   #if nothing happens, it succeeded, if failed it will return an exception error

#save file to hardrive
playFile = open('RomeoAndJuliet.txt', 'wb')   #wb means write binary mode - need to write binary to maintain the unicode 
for chunk in res.iter_content(100000):    #will return chunks in bytes over each iteration and can specify how many bytes per chunk to be returned
    playFile.write(chunk)   #this should have returned how many bytes per chunk but it didn't return anything?

playFile.close()

#it did still save the text file to the drive 
```
