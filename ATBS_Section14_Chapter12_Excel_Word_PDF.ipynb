```python
#chapter 12/section 14: video 44: Reading and Editing PDFs (pages 295 - 302 )

#pdfs are binary files so more complex than plaintext 

#pip install PyPDF2 - be sure to get the capitalisation correct! 

#might not be open some PDF's or extract some text but this cannot be helped, apparently PyPDF2 is the best one, it can extract text and return it as a python string, but cannot deal with pictures and charts 

# download example files from http://autbor.com/meetingminutes1.pdf -- and -- meetingminutes2.pdf 
import PyPDF2
import os
os.chdir('c:\\Users\\DHOPE\\Downloads') #if you don't set working directory would have to set the full path in the open 

pdfFile = open('ATBSSECTION14meetingminutes1.pdf', 'rb') #open in read binary mode 
reader = PyPDF2.PdfFileReader(pdfFile)
reader.numPages #how many pages are in the pdf doc

page = reader.getPage(0) #pages start at 0, not 1 
page.extractText()  #didn't extract it perfectly but mostly matches the text with the pdf , returned OOFFFFIICCIIAALL  BBOOAARRDD  MMIINNUUTTEESS but most other words were correct, this section of text was formatted differently which might be why it was harder to extract

#to get all text in the example pdf
#for pageNum in range (reader.numPages) : 
 #   print(reader.getPage(pageNum).extractText())
    
#to combine two pdf files into one

pdf1File = open('ATBSSECTION14meetingminutes1.pdf', 'rb')
pdf2File = open('ATBSSECTION14meetingminutes2.pdf', 'rb')

reader1 = PyPDF2.PdfFileReader(pdf1File)
reader2 = PyPDF2.PdfFileReader(pdf2File)

writer = PyPDF2.PdfFileWriter() #the new file that will contain the combined two pdfs - this isn't saved yet
for pageNum in range(reader1.numPages):
    page = reader1.getPage(pageNum)
    writer.addPage(page)

for pageNum in range(reader2.numPages):
    page = reader2.getPage(pageNum)
    writer.addPage(page)
#save to the drive the new pdf 
outputFile = open('combinedminutes.pdf', 'wb')
writer.write(outputFile)
#close all open pdfs
pdf1File.close()
pdf2File.close()
outputFile.close()

#the new file is a version that looks like you have combined using adobe ! 
```

```python
#chapter 12/section 14: video 45: Reading and Editing Word Documents (pages 233 - 236 )

#pip install python-docx 
#word docs have a lot of structure 

#download example working file from http://autbor.com/demo.docx

import docx
d = docx.Document('c:\\Users\\DHOPE\\Downloads\\demo.docx')

#look at first paragraph object at index 0
d.paragraphs[0].text #returns the text in this paragraph
d.paragraphs[1].text
p = d.paragraphs[1]

#a new run starts whenever there is a change in the style 
p.runs
p.runs[0].text
p.runs[1].text
p.runs[2].text
p.runs[3].text

p.runs[1].bold #returns true because this run is formatted as bold
p.runs[2].bold == None  #returns nothing because this run is formatted as italics 

#can change the format
p.runs[3].italic #this is true because it is already italic 
p.runs[3].underline = True   #set this section to be underlined

p.runs[3].text = 'italic and underlined.' #change the text in this section from just 'italic' to 'italic and underlined'
d.save('c:\\Users\\DHOPE\\Downloads\\demo2.docx') #save as a new file

#can also change the style of the run

p.style = 'Title'
d.save('c:\\Users\\DHOPE\\Downloads\\demo3.docx')

#create a brand new blank document that is nowhere in the hardrive 

d = docx.Document()
d.add_paragraph('Hello this is a paragraph')
d.add_paragraph('This is another paragraph')
d.save('c:\\Users\\DHOPE\\Downloads\\demo4.docx')

p = d.paragraphs[0]
p.add_run('This is a new run') #adds a new line in this paragraph 

p.runs
p.runs[1].bold = True #bolds this paragraph 
d.save('c:\\Users\\DHOPE\\Downloads\\demo5.docx')

#NOTE can only insert new paragraphs and runs at the end of what is already there 

#get all the text inside a word doc inside a string 

import docx

def getText(filename):
    doc = docx.Document(filename)
    fullText = []
    for para in doc.paragraphs:
        fullText.append(para.text)
    return '\n'.join(fullText)

print(getText('c:\\Users\\DHOPE\\Downloads\\demo.docx'))
```
